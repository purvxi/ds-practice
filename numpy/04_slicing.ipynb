```python
import numpy as np
arr = np.arange(1,26).reshape(5,5)
print (arr)
```

```python
#Get [13, 14, 15] from row
arr[2,2:]
```

```python
#Get [11, 12, 13] from row
arr[2,:3]
```

```python
#Get every other element: [11, 13, 15]
arr[2,::2]
```

```python
#Reverse row entirely
arr[2,::-1]
```

```python
#Get the entire 3rd row (index 2) → [11,12,13,14,15]
arr[2,:]
```

```python
#Get the entire last column → [5,10,15,20,25]
arr[:,4]
```

```python
#Get the first two rows, all columns
arr[0:2,:]
```

```python
#Get just the single value 18
arr[3,2]
```

```python
#Extract this 2×2 block from the middle:
# [[13,14],[18,19]]
arr[2:4,2:4]
```

```python
#Extract the 3×3 block in the bottom-right corner
arr[2:,2:]
```

```python
#Get every other row and every other column
arr[::2,::2]
```

```python
#Get the last row using negative indexing
arr[-1]
```

```python
#Get the last two columns of the last two rows
arr[3:,3:]
```

```python
# Reverse the rows only (flip vertically), keeping column order intact
arr[::-1,0:]
```

```python
#Reverse the entire array (both rows and columns flipped)
arr[0:,::-1]
```

```python
sub = arr[0:2, 0:2]
print(sub)
sub[0,0]=999
print(arr[0,0])
```

```python
#iteration
for i in np.nditer(arr):
  print(i)
```

```python
#transpose
#np.transpose(arr)
arr.T
```

```python
#ravel : converts to 1D array
arr.ravel()
```
