```python
#stacking
import numpy as np
a1=np.arange(12).reshape(2,6)
a1
a2=np.arange(12,24).reshape(2,6)
a2
```

```python
#horizontal stacking
np.hstack((a1,a2))
```

```python
np.hstack((a1,a2,a1,a2))
```

```python
#vertical stacking
np.vstack((a1,a2))
```

```python
np.vstack((a1,a2,a2,a1))
```
